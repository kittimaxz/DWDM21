```python
data = os.path.join(path,'รายการยืมหนังสือผ่านUSL.csv')
print(data)
```

```python
data_USL = pd.read_csv(data)
data_USL
```

```python
data_USL.info()
```

```python
data_USL.isnull().any()
```

```python
time_morning = data_USL[(data_USL['time']>='06:00:00') & (data_USL['time']<='11:59:59')] 
time_morning
```

```python
time_afternoon = data_USL[(data_USL['time']>='12:00:00') & (data_USL['time']<='17:59:59')] 
time_afternoon
```

```python
time_evening = data_USL[(data_USL['time']>='18:00:00') & (data_USL['time']<='23:59:59')] 
time_evening
```

```python
time_midnight = data_USL[(data_USL['time']>='00:00:00') & (data_USL['time']<='05:59:59')] 
time_midnight
```

```python
book_of_day = data_USL.groupby('day').count().sort_values(by=['order_code'],ascending=False) 
book_of_day.iloc[:,:1]
```

```python
book_of_date = data_USL.groupby('date').count().sort_values(by=['order_code'],ascending=False) 
book_of_date.iloc[:,:1]
```

```python
book_of_time = data_USL.groupby('time').count().sort_values(by=['order_code'],ascending=False) 
book_of_time.iloc[:,:1]
```

```python
book_of_people = data_USL.groupby('apps_usertype_name').count().sort_values(by=['order_code'],ascending=False) 
book_of_people.iloc[:,:1]
```

```python
book_of_namebook = data_USL.groupby('book_title').count().sort_values(by=['order_code'],ascending=False) 
book_of_namebook.iloc[:,:1]
```

```python
book_of_location = data_USL.groupby('book_location').count().sort_values(by=['order_code'],ascending=False) 
book_of_location.iloc[:,:1]
```

```python
book_of_namepeople = data_USL.groupby('apps_firstname').count().sort_values(by=['order_code'],ascending=False) 
book_of_namepeople.iloc[:,:1]
```

```python
book_of_std = data_USL.groupby('std').count().sort_values(by=['order_code'],ascending=False) 
book_of_std.iloc[:,:1]
```
